```python
file = 'train_df_무음제거.csv'
df  = pd.read_csv(file, encoding='cp949')
#df.drop('symptomatic', axis=1, inplace=True)
print('Shape: ', df.shape)
df.head()
```

```python
df.columns = ['uuid', 'cough_detected', 'SNR', 'gender', 'man',
       'respiratory_condition', 'fever_muscle_pain', 'COVID-19', 'healthy',
       'symptomatic', '15세미만', '15세이상65세미만', '65세이상']
```

```python
df.head()
```

```python
df['15세이상65세미만'].value_counts()
```

```python
# 15세이상65세미만 데이터만 뽑기
df_1565 = df[df['15세이상65세미만'] == 1 ]
df_1565.reset_index(drop=True, inplace=True)
print('Shape: ', df_1565.shape)
df_1565.head()
```

```python
# 15세미만, 65세이상 데이터만 뽑기
df_15 = df[df['15세미만'] == 1 ]
df_65 = df[df['65세이상'] == 1 ]
df_15.reset_index(drop=True, inplace=True)
df_65.reset_index(drop=True, inplace=True)
print('Shape: ', df_15.shape)
display(df_15.head())
print('Shape: ', df_65.shape)
display(df_65.head())
```

```python
# 남-음성, 남-양성, 여-음성, 여-양성 비율 맞추기
```

```python
print('--------- 15세미만 ---------')
display(df_15.groupby(['man', 'COVID-19']).sum())
print('--------- 15세이상65세미만 ---------')
display(df_1565.groupby(['man', 'COVID-19']).sum())
print('--------- 65세이상 ---------')
display(df_65.groupby(['man', 'COVID-19']).sum())
```

```python
# 15세미만
# # 남-음성, 남-양성, 여-음성, 여-양성 --> 각 20개
df_15_m_n = df_15[(df_15['man'] == 1) & (df_15['COVID-19'] == 0)][:20] # 남-음성
df_15_m_p = df_15[(df_15['man'] == 1) & (df_15['COVID-19'] == 1)][:20] # 남-양성
df_15_f_n = df_15[(df_15['man'] == 0) & (df_15['COVID-19'] == 0)][:20] # 여-음성
df_15_f_p = df_15[(df_15['man'] == 0) & (df_15['COVID-19'] == 1)][:20] # 여-양성
```

```python
# 65세 이상 
# # 남-음성, 남-양성, 여-음성, 여-양성 --> 각 13개
df_65_m_n = df_65[(df_65['man'] == 1) & (df_65['COVID-19'] == 0)][:13] # 남-음성
df_65_m_p = df_65[(df_65['man'] == 1) & (df_65['COVID-19'] == 1)][:13] # 남-양성
df_65_f_n = df_65[(df_65['man'] == 0) & (df_65['COVID-19'] == 0)][:13] # 여-음성
df_65_f_p = df_65[(df_65['man'] == 0) & (df_65['COVID-19'] == 1)][:13] # 여-양성
```

```python
# 데이터 병합 - 15세미만, 65세 이상 
df_15_65 = pd.concat([df_15_m_n, df_15_m_p, df_15_f_n, df_15_f_p ,df_65_m_n, df_65_m_p, df_65_f_n, df_65_f_p ])
df_15_65.reset_index(drop=True, inplace=True)
```

```python
print('Shape: ', df_15_65.shape)
df_15_65.head()
```

```python
# # 남-음성, 남-양성, 여-음성, 여-양성 --> 각 150개로 개수 및 비율 맞추기
df_1565_m_n = df_1565[(df_1565['man'] == 1) & (df_1565['COVID-19'] == 0)][:150] # 남-음성
df_1565_m_p = df_1565[(df_1565['man'] == 1) & (df_1565['COVID-19'] == 1)][:150] # 남-양성
df_1565_f_n = df_1565[(df_1565['man'] == 0) & (df_1565['COVID-19'] == 0)][:150] # 여-음성
df_1565_f_p = df_1565[(df_1565['man'] == 0) & (df_1565['COVID-19'] == 1)][:150] # 여-양성
```

```python
# 데이터 병합 - 15세이상65세미만
df_1565_150 = pd.concat([df_1565_m_n, df_1565_m_p, df_1565_f_n, df_1565_f_p])
df_1565_150.reset_index(drop=True, inplace=True)
```

```python
print('Shape: ', df_1565_150.shape)
df_1565_150.head()
```

```python
# csv 파일로 저장 
df_15_65.to_csv('train_df_무음제거_15세미만_65세이상.csv',index=False)
df_1565_150.to_csv('train_df_무음제거_15세이상65세미만.csv',index=False)
```
