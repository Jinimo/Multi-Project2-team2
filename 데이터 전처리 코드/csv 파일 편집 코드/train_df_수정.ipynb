```python
df  = pd.read_csv(file)
```

```python
df.columns
```

```python
df_n = df[['uuid', 'datetime', 'cough_detected', 'SNR', 'age', 'gender', 'respiratory_condition', 'fever_muscle_pain', 'status' ]]
```

```python
df_n.isna().sum()
```

```python
# 결측지 제거
df_n = df_n.dropna(axis=0)
```

```python
df_n = df_n.reset_index(drop=True)
```

```python
df_n.head()
```

```python
df_n.isna().sum()
```

```python
df_n2 = pd.concat([df_n, pd.get_dummies(df_n.status)], axis=1)
df_n2.head()
```

```python
df_n2[['respiratory_condition', 'fever_muscle_pain' ]] = df_n2[['respiratory_condition', 'fever_muscle_pain' ]].astype(int)
df_n2['age'] = df_n2['age'].astype(int)
```

```python
df_n2.head()
```

```python
df_n2['age_range'] = df_n2['age']
```

```python
def age_range_f(row):
    if row < 15:
        return '15세미만'
    elif 15 <= row and row < 65:
      return '15세이상65세미만'
    else:
        return '65세미만'
        
df_n2.age_range = df_n2.age_range.apply(age_range_f) 
```

```python
df_n2 = pd.concat([df_n2, pd.get_dummies(df_n2.age_range)], axis=1)
df_n2.head()
```

```python
data_loc=df_n2[['uuid', 'cough_detected', 'SNR', 'gender',
       'respiratory_condition', 'fever_muscle_pain', 'COVID-19', 
      'healthy', 'symptomatic', '15세미만', '15세이상65세미만', '65세미만']]
```

```python
# 남성이면 1, 여성이면 0의 값을 가지는 man column 생성
gender=[1 if data == 'male' else 0 for data in data_loc['gender'] ]
data_loc['man']=gender
```

```python
# 필요한 column만 뽑기
data_loc=data_loc[['uuid', 'cough_detected', 'SNR', 'gender', 'man', 'respiratory_condition',
       'fever_muscle_pain', 'COVID-19', 'healthy', 'symptomatic', '15세미만',
       '15세이상65세미만', '65세미만']]
```

```python
data_loc.head()
```

```python
# gender 항목의 other 제거
data_loc=data_loc[data_loc['gender']!='other']
```
