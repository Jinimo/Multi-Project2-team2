```python
import glob 
from os import path
import os
from pydub import AudioSegment
from pydub.utils import make_chunks

filepath='C:/Users/CJ/Desktop/MultiCampus/프로젝트_2_ai서비스 개발/data/public_dataset/'
lst = glob.glob(filepath + ".json")
for name in lst:
    os.remove(name)
```

```python

filelist=os.listdir(filepath)

for name in filelist:
    newname=name.split(".")[0]
    audioSegment = AudioSegment.from_file(filepath+name)
    audioSegment.export('./'+newname+'.wav', format='wav')
```
