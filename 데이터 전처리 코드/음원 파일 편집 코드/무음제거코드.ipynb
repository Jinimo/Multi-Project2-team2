```python
import librosa
import numpy as np
import pandas as pd
from scipy.io import wavfile
from librosa import display
import noisereduce as nr
import IPython.display as ipd
from tqdm.notebook import tqdm
```

```python
data_df=pd.read_csv('./train_df_kr.csv', encoding='cp949')
import os
volume=[]
location="C:/Users/User/Desktop/test1/"
filelist=data_df['uuid']
for filename in tqdm(filelist):
    data, sr=librosa.load(location+filename+".wav", sr=16000)
    volume.append(np.percentile(data, 95))
threshold=np.percentile(volume, 50)
```

```python
threshold
```

```python
def is_data_silence(data):
    filteredList = list(filter(lambda x: x>=threshold, data))
    if filteredList==[]:
        return True
    else:
        return False
```

```python
silence=[]
for filename in tqdm(filelist):
    data, sr=librosa.load(location+filename+".wav", sr=16000)
    if(is_data_silence(data)):
        silence.append(filename)

print(silence)
```

```python
len(silence)
```

```python
for i in silence:
    aa = data_df[data_df['uuid'].str.contains(i)].index
    data_df.drop(aa, inplace=True)
data_df
```
